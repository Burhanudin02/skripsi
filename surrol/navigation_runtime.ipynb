```python
import pybullet as p
import pkgutil, time
import torch

print("Attempting to connect to PyBullet with EGL")

try:
    cid = p.connect(p.DIRECT)
    
    egl = pkgutil.get_loader('eglRenderer')
    
    if egl:
        plugin = p.loadPlugin(egl.get_filename(), "_eglRendererPlugin")
        print(f"Successfully loaded EGL plugin: {plugin}")
        
    else:
        print("Could not find eglRenderer, rendering might fail.")
        
    print(f"torch version cuda: {torch.version.cuda}")
    
except Exception as e:
    print("\n-- EGL TEST FAILED --")
    print(f"An error occured: {e}")
    
finally:
    if 'cid' in locals() and p.isConnected(cid):
        p.disconnect(cid)
    print("Test finished.")
```
